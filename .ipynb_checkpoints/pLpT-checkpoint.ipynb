```python
import numpy as np
import matplotlib.pyplot as plt

data=np.array([])
filename='chi2cov_228.dat'

with open('MINERvA_nu/chi2Values/'+filename) as file:
    data = [[float(digit) for digit in line.split()] for line in file]
    
fig, ax = plt.subplots()
im = ax.imshow(data, origin="lower",extent=(1,1200,1,1200))

cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel(" ", rotation=-90, va="bottom")

ax.set_ylabel('$\omega$')
ax.set_xlabel('$q$')


fig = plt.figure(num=1, figsize=(1, 1), dpi=80, facecolor='w', edgecolor='k')
fig.savefig('pTpL.pdf', dpi=80, bbox_inches='tight', pad_inches=0)

plt.show()
```
